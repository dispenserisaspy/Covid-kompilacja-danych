```python
#Importuję porzebne biblioteki
import pandas as pd
import matplotlib.pyplot as plt

#Wczytuję dane z pliku excel
df = pd.read_excel(r"Python Projekt 2 - Covid19.xlsx")
#Zmieniam nazwy kolumn
df.rename(columns = {"dateRep": "Data","cases":"Zakażenia","deaths":"Zgony",
                     "countriesAndTerritories":"Kraj","continentExp":"Kontynent"}, inplace = True)

#================================================================================
#Funkcja do wyświetlania zakażeń i zgonów dla danego kraju w czasie
def Sortowanie_Kraj(kraj):
    #Uwaga! kraj =/= Kraj
    df_Kraj = df[df['Kraj'] == kraj].sort_values(by=['Data'], ascending=True)
    #Wyświetlam tabelkę z danymi
    display(df_Kraj)
    #Wielkość wykresu
    plt.figure(figsize=(15,5))
    plt.xlabel("Data")
    plt.ylabel("Liczba przypadków")
    plt.title("Wykres liczby zgonów i zakażeń dla danego kraju w czasie")
    #Tworzę wykres gdzie x to data a y to zgony
    plt.plot(df_Kraj.Data, df_Kraj.Zgony, color='r', marker='d', label='Zgony')
    #Tworzę wykres gdzie x to data a y to zakażenia
    plt.plot(df_Kraj.Data, df_Kraj.Zakażenia, color='b', marker='d', label='Zakażenia')
    plt.legend()
    #Dodaję siatkę
    plt.grid()

#Funkcja do wyświetlania globalnej sumy zakażeń i zgodnów w danym przedziale czasu
def Sortowanie_Czas(od,do):
    
    #To są pogrupowane po datach obserwacje
    df_zgrupowany = df.groupby('Data').sum().drop("day", axis=1).drop("month", axis=1).drop("year", axis=1).drop("popData2018", axis=1)
    
    #To jest kolumna zawierająca wyłącznie daty. 
    #Wyfiltrowane hardcode przez włochy ponieważ miały obserwacje przypisane do każdej daty w odróżnieniu do np. Polski
    df_daty = df[df['Kraj'] == 'Italy'].sort_values(by=['Data'], ascending=True).drop("Zakażenia", axis=1).drop("Zgony", axis=1).drop("Kraj", axis=1).drop("Kontynent", axis=1).drop("geoId", axis=1).drop("countryterritoryCode", axis=1).drop("day", axis=1).drop("month", axis=1).drop("year", axis=1).drop("popData2018", axis=1)

    #Indexy celowo resetowane bo inaczej metoda concat poniżej się rozjerzdża i zwraca mnóstwo NaN.
    df_zgrupowany.reset_index(drop=True, inplace=True)
    df_daty.reset_index(drop=True, inplace=True)
    
    #To jest suma logiczna powyższych dataframe'ów.
    df_wykres = pd.concat([df_zgrupowany, df_daty], axis=1)
    #display(df_wykres)
    
    #Tworzę dataframe z danymi z wybranego przedziału czasowego
    df_przedział = df_wykres[((df_wykres['Data'] <= do) & (df_wykres['Data'] >= od))]#.sort_values(by=['Data'], ascending=True)
    display(df_przedział)
    
    plt.figure(figsize=(15,5))
    plt.xlabel("Data")
    plt.ylabel("Liczba przypadków")
    plt.title("Wykres globalnej liczby zakażeń i zgonów w przedziale czasu")
    plt.plot(df_przedział.Data, df_przedział.Zgony, color='r', marker='d', label='Zgony')
    plt.plot(df_przedział.Data, df_przedział.Zakażenia, color='b', marker='d', label='Zakażenia')
    plt.legend()
    plt.grid()
    #display(df_wykres)

#================================================================================
#Interfejs dla użytkownika
print('Dzień dobry. \n Program daje możliwość sporządzenia wykresu zachorowań i zgonów w czasie dla danego kraju, \n lub globalny wzrost zachorowań i zgonów w wybranym przedziale czasowym ')
print('Proszę wybierz który z wariantów wybierasz: \n A: Zachorowania/Zgony dla Kraju. \n B: Globalne Zachorowania/Zgony w przedziale czasu.')
wariant = input()
if wariant == 'A':
    print('Wybrałeś wariant A')
    
    #Niestety Krajów jest zbyt dużo aby listować je użytkownikowi
    #lista_krajów = df["Kraj"].unique()
    #lista_krajów
    
    print('Proszę podać kraj dla którego chcesz otrzymać dane: \n UWAGA! Nazwy powinny być w języku angielskim a spacje zastąpione podłogą')
    
    państwo = input()
    if państwo in df.Kraj.values:
        Sortowanie_Kraj(państwo)
    else:
        print('Podano błędną nazwę kraju')

elif wariant == 'B':
    print('Wybrałeś wariant B')
    
    print('Proszę podać przedział czasu dla którego chcesz otrzymać dane: \n UWAGA! Daty powinny być wprowadzane w formacie RRRR-MM-DD')
    start = input()
    stop = input()
    Sortowanie_Czas(start,stop)
    
else:
    print('Nie ma takiej opcji')
```
